```python
import pandas as pd
from sqlalchemy import create_engine

def upload_dataframe_to_postgresql(df, table_name, db_url):
    """
    Uploads a DataFrame to a PostgreSQL table.

    Args:
        df (pd.DataFrame): The DataFrame to upload.
        table_name (str): The name of the PostgreSQL table.
        db_url (str): The database connection URL.

    Returns:
        bool: True if the upload was successful, False otherwise.
    """
    try:
        # Create a database engine
        engine = create_engine(db_url)

        # Upload the DataFrame to the specified table
        df.to_sql(table_name, engine, if_exists='replace', index=False)

        # Close the connection
        engine.dispose()

        print(f"DataFrame successfully uploaded to '{table_name}' in PostgreSQL.")
        return True
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return False
```

```python
from sqlalchemy import create_engine
import pandas as pd
from src.components.variable import dataBase

def fetch_data_using_sqlalchemy(table_name):
    try:
        db_url = dataBase.db_url
        engine = create_engine(db_url)

        query = f"SELECT * FROM {table_name}"
        df = pd.read_sql(query, engine)

        # Process the data as needed
        return df  # Print the first few rows as an example

    except Exception as e:
        print(f"An error occurred: {str(e)}")
    finally:
        engine.dispose()


table_name = "merge_table"
df=fetch_data_using_sqlalchemy(table_name)
```
